```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import os
import time

def chrome_preferences(dfolder, download=False):
    '''
    Sets the preferences of the Chrome browser: download path.
    '''
    # Create ChromeOptions object
    options = Options()
    
    # Set download preferences
    prefs = {
        "download.default_directory": dfolder,  # Set the default download directory
        "download.prompt_for_download": False,   # Disable the prompt for downloads
        "download.directory_upgrade": True,      # Ensure that downloads are saved to the specified folder
        "safebrowsing.enabled": True             # Enable safe browsing (optional)
    }
    options.add_experimental_option("prefs", prefs)

    # If the download flag is set, handle PDF downloads specifically
    if download:
        options.add_experimental_option("prefs", {
            "download.default_directory": dfolder,
            "download.prompt_for_download": False,
            "download.directory_upgrade": True,
            "safebrowsing.enabled": True,
            "plugins.always_open_pdf_externally": True  # Avoid opening PDFs in the browser
        })

    return options

def start_up(link, dfolder, geko_path, download=True):
    '''
    Initializes the Chrome browser with specified preferences and opens the provided link.
    '''
    os.makedirs(dfolder, exist_ok=True)  # Ensure the download folder exists

    options = chrome_preferences(dfolder, download)
    service = Service(geko_path)
    driver = webdriver.Chrome(service=service, options=options)

    # Open the specified link
    driver.get(link)
    time.sleep(5)  # Adjust sleep time as needed

    return driver

def check_and_click(browser, xpath, type):
    '''
    Function that checks whether the object is clickable and, if so, clicks on
    it. If not, waits one second and tries again.
    '''
    ck = False
    ss = 0
    while ck == False:
        ck = check_obscures(browser, xpath, type)
        time.sleep(1)
        ss += 1
        if ss == 15:
            # warn_sound()
            # return NoSuchElementException
            ck = True
            # browser.quit()

def check_obscures(browser, xpath, type):
    '''
    Function that checks whether the object is being "obscured" by any element so
    that it is not clickable. Important: if True, the object is going to be clicked!
    '''
    try:
        if type == "xpath":
            browser.find_element('xpath',xpath).click()
        elif type == "id":
            browser.find_element('id',xpath).click()
        elif type == "css":
            browser.find_element('css selector',xpath).click()
        elif type == "class":
            browser.find_element('class name',xpath).click()
        elif type == "link":
            browser.find_element('link text',xpath).click()
    except (ElementClickInterceptedException, NoSuchElementException, StaleElementReferenceException) as e:
        print(e)
        return False
    return True
```

```python
#Function string_to_date_period to serve as input in the df
import pandas as pd

def string_to_date_period(date_list):
    # Convert the first and last dates in the list to datetime objects
    start_date = pd.to_datetime(date_list[0])
    end_date = pd.to_datetime(date_list[-1])
    
    # Format the start and end dates as strings
    start_date_str = start_date.strftime('%Y-%m-%d')
    end_date_str = end_date.strftime('%Y-%m-%d')
    
    # Create the date period string
    date_period = f"{start_date_str} : {end_date_str}"
    return date_period


# Call the function and get the date period
date_period = string_to_date_period(date_list_may)

# Place the date period in the first row under "Date Period"
df.loc[0, 'Date Period'] = date_period

# Drop rows where both "City" and "Date Period" are NaN
df.dropna(how='all', inplace=True)

# Print the updated DataFrame
print(df)
```

```python
import time
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException

def scroll_and_load_more(driver, max_scrolls=20):
    scroll_attempts = 0  # Count how many times we scroll without finding new content

    while scroll_attempts < max_scrolls:
        try:
            # Check if the "Load More" button is visible and clickable
            load_more_button = driver.find_element(By.XPATH, '//button[span[text()="Cargar más resultados"]]')
            driver.execute_script("arguments[0].scrollIntoView();", load_more_button)

            # Try clicking with retry logic
            for _ in range(3):  
                try:
                    load_more_button.click()
                    print("Clicked 'Cargar más resultados' button")
                    time.sleep(3)  # Allow time for content to load
                    scroll_attempts = 0  # Reset scroll counter if we clicked successfully
                    break  
                except ElementClickInterceptedException:
                    print("Click was intercepted, retrying...")
                    time.sleep(2)
            else:
                print("Failed to click button after multiple attempts.")

        except NoSuchElementException:
            print("No 'Load More' button found. Trying to trigger lazy loading...")

            last_height = driver.execute_script("return document.body.scrollHeight")

            # **New scrolling method: Scroll in small steps instead of jumping to the bottom**
            for _ in range(3):  # Try scrolling a few times to load more content
                driver.execute_script("window.scrollBy(0, 500);")  # Scroll down slightly
                time.sleep(2)  # Wait for potential new elements
                new_height = driver.execute_script("return document.body.scrollHeight")

                if new_height == last_height:  # If scrolling does not change height, increase counter
                    scroll_attempts += 1
                else:
                    scroll_attempts = 0  # Reset scroll counter if new content loads
                    break  

            if scroll_attempts >= max_scrolls:
                print("No more content loading. Exiting.")
                break  # Exit the loop if we have scrolled too many times without new content

    print("Reached the bottom of the page or exhausted attempts.")

scroll_and_load_more(driver)
```

```python
hotels_data = driver.find_elements(By.XPATH, '//div[@data-testid="property-card"]')

hotels_list = []
ratings_list = []

for hotel in hotels_data:
    # Extract hotel name
    try:
        hotel_name = hotel.find_element(By.XPATH, './/div[@data-testid="title"]').text
    except NoSuchElementException:
        hotel_name = "Unknown"

    # Extract rating
    try:
        rating = hotel.find_element('xpath', './/div[@data-testid="review-score"]').text.split()[1]
    except (NoSuchElementException, IndexError):
        rating = 0
    print(f"{hotel_name} and {rating}")
    hotels_list.append(hotel_name)
    ratings_list.append(rating)

print(f"Extracted {len(hotels_list)} hotels and {len(ratings_list)} ratings.")
```

```python
#The DataFrame should have the same length that the number of hotels:
df = df.reindex(df.index.repeat(len(hotels_list)))
```

```python
# Reset index to ensure a clean, sequential index
df.reset_index(drop=True, inplace=True)
```

```python
#Now we should retrieve the prices
prices = driver.find_elements(By.XPATH, '//span[@data-testid="price-and-discounted-price"]')
for index, price_element in enumerate(prices, start=1):
    price = price_element.text  # Extract the price text (e.g., "€ 319")
    print(f"{index}. {price}")

prices_list = []
for price_element in prices:
    price_text = price_element.text 
    prices_list.append(price_text)

numeric_prices = []

for price in prices_list:
    numeric_price = float(price.replace('€', '').replace('.', '').strip())
    numeric_prices.append(numeric_price)

# Print the numeric list of prices
print("Numeric Prices List:", numeric_prices)

```
